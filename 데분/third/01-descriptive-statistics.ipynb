```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

pd.set_option("display.max_columns", 100)
sns.set_style("whitegrid")
```

# Data Loading

[Ref] https://en.wikipedia.org/wiki/Iris_flower_data_set

```python
_data = load_iris() #_로 데이터 프레임의 완성정도를 표현 _a -> a -> a_ 수
_data.keys()
```

```python
data = _data['data']
feature_names = _data['feature_names']

df = pd.DataFrame(data, columns=feature_names)
df['target'] = _data['target']

df.head()
```

# Preprocessing

```python
df.describe()
```

교수님은 mean, std 먼저 봄. -> 평균, 표준편차가 대채로 작음.  

target은 min == 25%, max ==75% -> 이를 통해 범주형임을 추론할 수 있음

```python
# target 재정의
df['target'] = df['target'].map(
    {
        0: 'Iris-Setosa',
        1: 'Iris-Versicolour',
        2: 'Iris-Virginica',
    }
)

df.head()
```

```python
a1 = df[df['target'] == 'Iris-Setosa'].drop('target', axis=1)
a2 = df[df['target'] == 'Iris-Versicolour'].drop('target', axis=1)
a3 = df[df['target'] == 'Iris-Virginica'].drop('target', axis=1)
```

```python
def zscore(df):
    return ((df - df.mean())/df.std())

for col in a1:
    print(col)
    sns.kdeplot(zscore(a1[col]))
    sns.kdeplot(zscore(a2[col]))
    sns.kdeplot(zscore(a3[col]))
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 변수 설정
features = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]

# 종별 표준화 수행
df_scaled = df.copy()
species = df["target"].unique()

for specie in species:
    scaler = StandardScaler()
    mask = df["target"] == specie
    df_scaled.loc[mask, features] = scaler.fit_transform(df.loc[mask, features])

# 색상 설정
colors = {"Iris-Setosa": "blue", "Iris-Versicolour": "green", "Iris-Virginica": "red"}

# 플롯 생성
plt.figure(figsize=(10, 6))

for i, feature in enumerate(features):
    for j, specie in enumerate(species):
        values = df_scaled[df["target"] == specie][feature]
        x_positions = np.full_like(values, i + 1, dtype=float) + (j - 1) * 0.15  # 종별로 약간 옆으로 배치
        plt.scatter(x_positions, values, color=colors[specie], label=specie if i == 0 else "", alpha=0.1)

# x축 설정
plt.xticks(range(1, len(features) + 1), features, rotation=15)
plt.xlabel("Features")
plt.ylabel("Standardized Value (Within Each Species)")
plt.title("Standardized Feature Distribution by Species")

# 범례 추가
plt.legend(title="Species")
plt.grid(axis='y', linestyle='--', alpha=0.6)

# 플롯 표시
plt.show()

```

### skew

```python
_df_skew = df.groupby('target').skew()

_df_skew.T.plot(kind='bar', grid=True)
plt.title('Skewness by Feature with IRIS')
plt.ylabel('Skewness')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### kurtosis

```python
df.groupby('target').kurtosis()  # 뾰족한 정도 
```

```python
df.groupby("target").apply(
    lambda group: group.iloc[:, :-1].kurtosis()
)
```

```python
_df_kurt = df.groupby("target").apply(lambda group: group.iloc[:, :-1].kurtosis())

_df_kurt.T.plot(kind='bar', grid=True)
plt.title('Kurtosis by Feature with IRIS')
plt.ylabel('kurtosis')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
